**Menampilkan Multiplot**

```python
kolom_pilih = ['953', '915', '989', '957', '942', '945']
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

for i, kolom in enumerate(kolom_pilih):
    baris_tertentu = data2.loc[:3599, kolom]
    baris_tertentu.plot(ax=axs[i//3, i%3], title=kolom)

plt.tight_layout()
plt.show()
```

**Plot scatterplot rata-rata**

```python
rata_rata = data2.loc[:3599].mean()

warna = np.arange(len(rata_rata)) // 10

plt.figure(figsize=(10, 6))
for kelompok in np.unique(warna):
    plt.scatter(rata_rata.index[warna == kelompok],
                rata_rata[warna == kelompok],
                label=f'Kelompok {kelompok + 1}')

plt.xlabel('Kolom')
plt.ylabel('Nilai Rata-Rata')
plt.title('Scatter Plot Nilai Rata-Rata Setiap Kolom')
plt.legend()

plt.grid(True)
plt.tight_layout()
plt.show()
```

**Plot scatterplot skewness**

```python
skewness = data2.loc[:3599].apply(lambda x: x.skew())

warna_skew = np.arange(len(skewness)) // 10

plt.figure(figsize=(10, 6))
for kelompok in np.unique(warna_skew):
    plt.scatter(skewness.index[warna_skew == kelompok],
                skewness[warna_skew == kelompok],
                label=f'Kelompok {kelompok + 1}')

plt.xlabel('Kolom')
plt.ylabel('Skewness')
plt.title('Scatter Plot Skewness Setiap Kolom')
plt.legend()

plt.grid(True)
plt.tight_layout()
plt.show()
```
