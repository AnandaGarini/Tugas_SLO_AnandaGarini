**4 Player pencetak goal terbanyak dari MNC**

```python
club_syarat = 'Manchester City'
pemainmnc = data[data['team_title'] == club_syarat]

mostgoalmnc = pemainmnc.sort_values(by='goals', ascending=False).head(10)
mostgoalmnc
```

**Dari kesebelasan mana pencetak goal terbanyak**

```python
club = data.groupby('team_title')['goals'].sum().reset_index()
terbaikbasedgoal = club.loc[club['goals'].idxmax()]
print(f"\nClub dengan goal terbanyak:")
print(terbaikbasedgoal)
```

**Berapa jumlah kartu kuning dan merah yang diterima pembuat asssist terbanyak**

```python
most_assist = data['assists'].max()
pemain_assist = data[data['assists'] == most_assist]
show_pemain = pemain_assist[['player_name', 'assists', 'yellow_cards','red_cards']]
show_pemain
```

**Siapa yang jumlah assist dan goalnya paling banyak dan dari kesebelasan mana?**

```python
most_assist_2 = data['assists'].max()
pemain_assist_2 = data[data['assists'] == most_assist_2]
show_pemain = pemain_assist_2[['player_name', 'assists', 'team_title']]
show_pemain
```

```python
most_goal_2 = data['goals'].max()
pemain_goal_2 = data[data['goals'] == most_goal_2]
show_pemain_goal = pemain_goal_2[['player_name', 'goals', 'team_title']]
show_pemain_goal
```

```python
most_most_again = data[(data['goals'] == most_goal_2) & (data['assists'] == most_assist_2)]
most_most_player = most_most_again[['player_name', 'assists', 'team_title']]
most_most_player
```

**5 Kesebelasan yang menyumbang goal terbanyak**

```python
club_5 = data.groupby('team_title')['goals'].sum().reset_index()
terbaikbasedgoal_5 = club_5.sort_values(by='goals', ascending=False)
again = terbaikbasedgoal_5.head(5)
again
```
